```python
import pandas as pd
import numpy as np
import os
```

```python
global_pathway = '/home/antton/TFM/GWAS/output/May27_2021/CB_May27_B-NK-CD38+.txt'

f = pd.read_csv(global_pathway, sep="\t", chunksize=100000)
list_of_downsampled_chunks=[]
for chunk in f:

    list_of_downsampled_chunks.append(chunk[chunk['pval']<1.0e-06].copy())
```

```python
#Isolate Phenotype name from path
phenotype = ''.join(global_pathway.split('/')[-1].split('.')[0].split('_')[2:])
```

NOTE: Coordinate system is 'hg19'

```python
concat_df = pd.concat(list_of_downsampled_chunks)
concat_df = concat_df.assign(Phenotype=phenotype)
concat_df = concat_df[concat_df['pval'] <= 5e-8]  # additional pvalue filter
concat_df
```

```python
output_df = concat_df.copy()
```

```python
#Rename column names
COLname_list = output_df.columns.tolist()
COLname_list[2] = 'hg37 position'
output_df.columns = COLname_list
```
